### 2. Data preprocessing

### Processing missing values

```python
nantypes = df[df.isnull().any(axis=1)]
nantypes['debt'].value_counts()
```

```python
df.isna().sum()
```

```python
df["children"].replace({20: 0, -1: 0}, inplace=True)
```

We have replaced some of the errors in the data with values we could predict or seemed logical.<br>
For example we had 76 entries in the 'kids' column which values were '20', As this seems as a typo (Either users entered '0' or '2').<br>
Same goes for entries in that particular column which are valued as '-1', We will convert all of the oddities to the '0' value which is our median value.

The missing or wrong values were present probably because of an input mistake(typing a string value to integer/float field or miss clicks of users when submitting the information (-1 kids, or 20 when they intended to type 2 or 0)

```python
df["children"].value_counts()
```

```python
# 101 entries of age = "0", since all the attributes are fine well replace with mean() or median() since they are about the same
df[
    df["dob_years"] < 18].count()
df["dob_years"].median() # 42
df["dob_years"].mean()  # 43.29337979094077

df["dob_years"] = df["dob_years"].replace(to_replace = 0, value= df["dob_years"].median())
```

```python
df[df["dob_years"] < 18].count()
```

```python
#fill na with median of income.
df['total_income'] = df['total_income'].fillna(df['total_income'].median())
df['days_employed'] = df['days_employed'] * -1
df['days_employed'] = df['days_employed'].fillna(df['days_employed'].median())
df['years_employed'] = df['days_employed'] / 365
```

```python
df.isna().sum()
```

Missing values were present simultaneously in 'years_employed' and 'total_income' columns and were filled with their respective medians.<br>
Negative and positive integeres were reversed.<br>
Also an 'years_employed' column was added for convenience.<br>

### Categorizing Data

```python
# we will sort a data frame with the relevant columns "children", "family_status_id", "debt", "total_income" and "purpose" 
# for profiling our customers.
df_cat = df[["children", "family_status_id", "debt", "total_income", "purpose"]]
print(df_cat.head(10))
```

Our new dataframe is cleaner and the only thing left is to sort the 'purpose' column.<br>
We will use lemmatizer and a stemmer to extract one of our four purposes: Education, Real estate, Vehicle purchase or Weddings/loans.<br>
Later we will assign a category value to each one of the loan purposes by ascending loan "size".

```python
#define nltk's
lmtzr = WordNetLemmatizer()
stemmer = SnowballStemmer('english') 

#first we tokenize the string in the column
df['tokenize'] = df.purpose.apply(lambda row: word_tokenize(row))
# we extract the short version of each purpose 
lst = ["house", "housing", "estate", "property","car","wedding", "education" , "educated", "university"]
def fun(a):
    for i in lst:
        if i in a:
            return i
    return "others"
df["category_test"] = df.tokenize.apply(lambda x: fun(x))


def category(cat):
    if cat == 'educ' or cat == 'univers':
        return "education"
    if cat == 'hous' or cat == 'estat' or cat == 'properti':
        return "house"
    if cat == 'wed':
        return "wedding"
    return "car"

#we narrow down the options and make them more unified    

df['test'] = df['category_test'].apply(lambda x: stemmer.stem(x))
df['category'] = df['test'].apply(lambda x: category(x))
```

```python
#now whats left is to procure a dataframe which is relevant to our assignment.
df2 = df[["children", "family_status_id", "debt", "total_income", "category"]]
df2.head(30)
```

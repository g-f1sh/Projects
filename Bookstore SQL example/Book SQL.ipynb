```python
query_2 = '''
SELECT
   ratings.book_id,
   COUNT(DISTINCT(review_id)) AS review_cnt,
   AVG(rating)

FROM
    ratings
    INNER JOIN reviews ON reviews.book_id = ratings.book_id
    
GROUP BY
    ratings.book_id
    
ORDER BY
    review_cnt DESC
'''


pd.io.sql.read_sql(query_2, con = engine)
```

<center><b>Identify the publisher that has released the greatest number of books with more than 50 pages</b></center>

```python
query_3 = '''
SELECT
    books.publisher_id,
    publishers.publisher,
    count(books.book_id)

FROM
   books
   INNER JOIN publishers ON publishers.publisher_id = books.publisher_id
    
WHERE
    books.num_pages > 50
GROUP BY
    books.publisher_id,
    publishers.publisher
ORDER BY
    count DESC
'''


pd.io.sql.read_sql(query_3, con = engine)
```

<center><b>Identify the author with the highest average book rating (look only at books with at least 50 ratings).</b></center>

```python
query_4 = '''
SELECT
    sub.author_id,
    sub.author,
    sub.avg,
    sub.cnt

FROM
    (SELECT 
        books.author_id,
        authors.author,
        AVG(rating) AS avg,
        COUNT(rating_id) AS cnt
    FROM 
        books
        INNER JOIN authors ON authors.author_id = books.author_id
        RIGHT JOIN ratings ON ratings.book_id = books.book_id
    GROUP BY
        books.author_id,
        authors.author
        ) AS sub
WHERE
    cnt > 50
ORDER BY
    avg DESC



'''


pd.io.sql.read_sql(query_4, con = engine)
```

<center><b>Find the average number of text reviews among users who rated more than 50 books.</b></center>

```python
query_5 = '''
SELECT
    AVG(reviews_count.text_count)
FROM
    (SELECT
        reviews.username,
        COUNT(reviews.text) AS text_count
    FROM
        reviews
    WHERE
        username IN (SELECT
            ratings.username
        FROM
            ratings
        GROUP BY
            ratings.username
        HAVING
            COUNT(book_id) >= 50)
    GROUP BY
            reviews.username) AS reviews_count
'''


pd.io.sql.read_sql(query_5, con = engine)
```

Well see how much of the data is missing:

```python
for (columnName, columnData) in df.iteritems():
    if columnData.isna().sum() != 0:
        print("{} has {} NaNs which is {:,.0f}% of the data".format(columnName, columnData.isna().sum(), columnData.isna().sum() / 51525 *100 ))

        
# or

df[df.columns[df.isnull().any()]].isnull().sum() * 100 / df.shape[0]
```


### Course of Action.
Right away as we can see that the easiest column to fix the "is_4wd",
as the missing values are clearly "1", there no other option.<br>
The values are Missing at Random which is a slight problem, "cylinders" and "paint_color" are necessary to answer future question and poses a big chunk of our data to simply delete it.
we'll fill the cylinders column with group by and the "ffill" method of fillna.<br>
before we proceed to the "odometer" and "model_year" we will deal with what we can fix first:

```python
df['is_4wd'].fillna('0', inplace=True)
df['is_4wd'] = df['is_4wd'].astype(int)
# and also lowercase the type column
df['type'] = df['type'].str.lower()
```

    
We would fill the cylinder column using groupby on the model column,
then use forward fill method in the fillna, this will replace all the values with a more dynamic range keeping the rows and only slightly changing the data volume.<br>
odometer will be filled with the mean by year groupoing including all the 0 values that dont make sense so we can avoid the big outliers affecting our data.<br>
model_year will be dropped since there is no logical way filling it without guessing (sure, we can roughly calculate the year by the odometer column. but as we artificially changed that column it would be biased)

```python
df.dropna(subset=["model_year"], inplace=True)
df["model_year"] = df["model_year"].astype(int)
```

```python
group_odometer = df.groupby('model_year')['odometer'].median()

for i in group_odometer.index:

    df.loc[(df['model_year'] == i) & (df['odometer'].isnull() | df['odometer'] == 0), 'odometer'] = group_odometer[i]
    

df['odometer'] = df['odometer'].fillna(df.groupby('model_year')['odometer'].transform('median'))
```

```python
df.dropna(subset=['odometer'], inplace=True)
df['odometer'] = df['odometer'].astype(int)
```

```python
df['paint_color'].fillna('other', inplace=True)
df['paint_color'] = df['paint_color'].astype(str)
```

```python
df['cylinders'] = df.groupby('model')['cylinders'].fillna(method='ffill')
df['cylinders'] = df['cylinders'].fillna(df['cylinders'].mode()[0])
df['cylinders'] = df['cylinders'].astype(int)
```

```python
def convert_condition(row):
    if row == "new":
        return 5
    if row == "like new":
        return 4
    if row == "excellent":
        return 3
    if row == "good":
        return 2
    if row == "fair":
        return 1
    if row == "salvage":
        return 0
```

```python
df['condition'] = df['condition'].apply(convert_condition)
```

### Step 3. Make calculations and add them to the table

```python
# we convert to datetime:
df['date_posted'] = pd.to_datetime(df['date_posted'], format='%Y-%m-%d')
df['day'] = df['date_posted'].dt.weekday
```

```python
df['month'] = df['date_posted'].dt.month
df['year'] = df['date_posted'].dt.year
df['model_age'] = df['year'] - df['model_year']
df['model_age'] = df['model_age'].astype(int)
```

```python
def avg_fun(row):
    age = row['model_age']
    odometer = row['odometer']
    try:
        return odometer / age
    except:
        return odometer
            
df['avg_mil'] = df.apply(avg_fun, axis = 1)
df['avg_mil'] = df['avg_mil'].astype(int)
```


### Conclusion
We added all the new columns,<br> there was an issue while creating the avg_mil col, as there were divisions by 0 which caused NaNs and infs to appear, so i changed all the 0 values to 1
